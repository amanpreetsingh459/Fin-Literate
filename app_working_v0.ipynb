```python
import gradio as gr

def quiz_app():
    # Define your questions and answers here
    questions = [
        {
            "question": "What is the capital of France?",
            "choices": ["Berlin", "Madrid", "Paris", "Rome"],
            "correct_answer": "Paris"
        },
        {
            "question": "Which planet is known as the Red Planet?",
            "choices": ["Mars", "Venus", "Jupiter", "Saturn"],
            "correct_answer": "Mars"
        },
        # Add more questions as needed
    ]

    def check_answer(choice, question_data):
        if choice == question_data["correct_answer"]:
            return "Correct!", 1  # Return "Correct!" and increment score by 1
        else:
            return "Incorrect.", 0  # Return "Incorrect." and no score increment

    with gr.Blocks() as demo:
        # State variables to track the quiz progress
        current_question = gr.State(0)
        score = gr.State(0)

        # UI elements
        gr.Markdown("## Multiple Choice Quiz")
        question_display = gr.Markdown(questions[0]["question"])  # Initial question
        choices = gr.Radio(questions[0]["choices"], label="Choices")
        submit_btn = gr.Button("Submit")
        output = gr.Markdown()

        # Event handling
        def on_submit(choice):
            result, score_increment = check_answer(choice, questions[current_question.value])
            score.value += score_increment

            # Move to the next question or end the quiz
            current_question.value += 1
            if current_question.value < len(questions):
                # Update UI elements directly, including choices
                return (
                    questions[current_question.value]["question"],
                    gr.update(choices=questions[current_question.value]["choices"]),  # Update choices here
                    result + f"\n Current score: {score.value}"
                )
            else:
                # Quiz finished
                submit_btn.disabled = True
                return (
                    "Quiz finished!",
                    gr.update(choices=[]),  # Clear choices when the quiz is finished
                    f"Your final score is {score.value}/{len(questions)}"
                )

        submit_btn.click(on_submit, inputs=choices, outputs=[question_display, choices, output])

    demo.launch()
```

```python
if __name__ == "__main__":
    quiz_app()
```
